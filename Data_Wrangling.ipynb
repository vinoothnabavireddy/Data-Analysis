### MISSING DATA & HANDLING
###### Some of our columns have missing values and to dentify which columns have missing values
###### We need to first convert our missing data in our dataset which is represented by a question mark using the replace method

```python
# To handle the missing values we will import the numpy library.
import numpy as np

#to create plots or graphs we need two libraries--
import matplotlib.pylab as pt

#To consider NaN (Not a numeric) as null values we need to set it using the following function.
pd.options.mode.use_inf_as_na = False
```

```python
# Step 1: Replacing "?" in our dataset and setting them to NaN.
data.replace('?', np.nan, inplace = True)
data
# If you would like to keep your original dataset remove inplace = True and save the result to a different variable.
# For example data1 = data.replace('?', np.nan)
#data1
```

```python
#Identifying columns with null values
missing_data = data.isnull() ##or you can also say dat.isna()
missing_data.head(10)

## The resulting data is a boolean value( True / False)
```

```python
#The above table makes it tough to determine which columns contain missing values and thus we can show a summary of missing values in each column by using the following function.
data.isnull().sum()
```

###### The result is a summary of total missing values in each column.

```python
# Another way of identifying columns with missing values is
missing_data.any()

#True value indicates that there is missing data in that column
```

```python
#Method 2 to display missing values using the boolean table is using a for loop
for i in missing_data.columns:
    print(i) #name of the column
    print(missing_data[i].value_counts()) # which gives you the number of flase or number of trues
    print("_______________________________________________________") # seperator for summary
```

##### Now we know the columns with missing values and the number of records that are to be handled.
##### For example: Normalized losses has 41 missing values,number of doors has 2 missing values and so on....
#### Techniques to handle missing values
##### 1. Droping the records entirely and resetting the index
##### 2. Replacing the missing values by either the mean or frequency.
##### 3. Using the min - max technique.

#### Replacing the NaN values in Normalized-losses column by the mean.
##### Since there are 41 records that have missing values which is a quarter of our dataset we cannot drop them. Hence, we will use the replace function to handle the missing values.


```python
# Calculate the mean of the column normalized-losses.
mean_normalized_losses = data["normalized-losses"].astype("float").mean(axis=0)
print("Mean of Normalized-losses is", mean_normalized_losses)
```

```python
#Replacing the NaN values with the calculated mean.
# It is important to note that data type conversion is required to replace the NaN values and hence we are using astype() to convert values to float.
data["normalized-losses"].replace(np.nan, mean_normalized_losses, inplace=True)
data.head(5)
```

```python
#The above step can be repeated for other columns with missing values
mean_bore = data['bore'].astype('float').mean(axis=0)
print("mean of bore is", mean_bore)
data["bore"].replace(np.nan, mean_bore, inplace=True)
mean_stroke = data['stroke'].astype('float').mean(axis=0)
print("mean of stroke is", mean_stroke)
data["stroke"].replace(np.nan, mean_stroke, inplace=True)
data.head(5)
```

```python
# For missing values in column num-of-doors
# unique values in the column are
data['num-of-doors'].unique()
```

##### So there are two unique values two doors and four doors.
##### To handle missing values in this column we should see the frequency of occurrence of four and two doors in the dataset.


```python
#To count the number of values of each we can use value_counts()
data['num-of-doors'].value_counts()
```

```python
# Since the most occurring value is four doors we will replace the NaN value with four.
data['num-of-doors'].replace(np.nan, 'four', inplace=True)
data.head(10)
```

```python
# For the column price we will drop the records with missing values and rest the index.
# This is because our aim is to predict the price so we will not replace NaN.
data.dropna(subset=["price"],axis=0, inplace=True)

#resetting index
data.reset_index(drop=True, inplace=True)

data.info()
```

##### The data has now been handled for all types of missing values

### Data Standardization

##### This primarily involves steps like converting data into a common format
##### example: Unit conversion from mpg to L/100Km 

```python
#The conversion from mpg to L/100 km is to divide 235/ mpg
data['highway-L/100km'] = 235/data['highway-mpg']
data['city-L/100km'] = 235/data['city-mpg']
data. head(3)
```

```python
#### Data Normalization
#### Involves transforming data into a similar range such as between 0 and 1.
#### This type will help with dealing with outliers.
data['length']=data["length"]/data["length"].max()
data['width']=data["width"]/data["width"].max()
data[['length','width']].head()
```

### Binning Data

```python
#data["horsepower"]= data["horsepower"].astype(int)
data.dtypes
data['horsepower'].unique()
```

```python
data.dropna(subset=["horsepower"],axis=0, inplace=True)
data['horsepower'].unique()
#resetting index
data.reset_index(drop=True, inplace=True)
data.head(3)
```

```python
data["horsepower"]=data["horsepower"].astype(int)
data.dtypes
```

```python
#Creating histogram to visualize the distribution of the data in the column horsepower.
%matplotlib inline
import matplotlib as pt
from matplotlib import pyplot
pt.pyplot.hist(data["horsepower"])

# To assign axis labels and title:
pt.pyplot.xlabel("horsepower")
pt.pyplot.ylabel("count of cars")
pt.pyplot.title("Distribution of Horsepower")
```

```python
#We can name the bins by creating a variable.
grouping =  ['Low HP', 'Medium HP','High HP']
print(grouping)
```

```python
#Using histogram to show binning
n= 3
pt.pyplot.hist(data["horsepower"], bins= n)
```

```python
#We can bin the data by:
bins = np.linspace(min(data['horsepower']),max(data['horsepower']), 4)
#Using the cut fucntion we cana ssign our groups to the respective bins based on the condition.
data['horsepower-binned']= pd.cut(data['horsepower'],bins,labels=grouping, include_lowest=True)
data[["horsepower","horsepower-binned"]].head(10)
```

```python
#To see how many cars are low medium or high HP. we can use the value_count function
data["horsepower-binned"].value_counts()
```

```python
# Bar graph to see visualize the above data
%matplotlib inline
import matplotlib as pt
from matplotlib import pyplot
pt.pyplot.bar(grouping,data["horsepower-binned"].value_counts())

pt.pyplot.xlabel("Horsepower")
pt.pyplot.ylabel("Number of Cars")
pt.pyplot.title("Horsepower after binning")
```
